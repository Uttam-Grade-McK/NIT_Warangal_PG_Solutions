# Import Required Libraries

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import os
os.chdir('D:\Edureka DataSets')
```

```python
# Read the file
df = pd.read_json('NewsArticles.json')
df
```

```python
# So we change the type the this Vector column because we cannot apply the KMeans and PCA on this single column.
det = df['Vector']
det
```

```python
# Make the data into DataFrame so every single column and rows can its different term
Vectors = pd.DataFrame(list(det))
Vectors
```

# Now Apply KMeans

```python
model = KMeans(n_clusters = 5)
model.fit(Vectors)
print('SSE:',model.inertia_)
```

# Now Apply PCA

```python
pca = PCA(n_components = 100)
pca.fit(Vectors.transpose())
new_vectors = pd.DataFrame(pca.components_).transpose()
new_vectors.shape
```

# KMean After PCA

```python
model_afterPCA = KMeans(n_clusters = 5)
model_afterPCA.fit(new_vectors)
print('SSE:',model_afterPCA.inertia_)
```

```python
df['labelwithoutpca'] = model.labels_
```

```python
df['labelafterpca'] = model_afterPCA.labels_
```

```python
for i,j in df.groupby('labelwithoutpca'):
    print('Label',i)
    print(' '.join(j['Preprocessed-Article']).lower().encode('utf-8').split()[:50])
```

```python
for i, j in df.groupby('labelafterpca'):
    print('Label',i)
    print(' '.join(j['Preprocessed-Article']).lower().encode('utf-8').split()[:10])
    df['labelwithoutpca'].value_counts()
df['labelafterpca'].value_counts()
```
